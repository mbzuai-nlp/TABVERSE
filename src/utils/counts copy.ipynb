```python
import json

path ='../data/2-task/totto.json'
with open('../data/2-task/totto.json') as f:
    data = json.load(f)
```

```python
data[0]
```

```python
# rename the id key as image_id and id key again and start from 0 to n
for i, item in enumerate(data):
    item['image_id'] = item.pop('id')
    item['id'] = i
```

```python
# order keys as follow
order = ['id', 'image_id', 'title', 'context', 'table', 'query', 'label']
data = [{k: item[k] for k in order} for item in data]
```

```python
data[0]
```

```python
with open(path, 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=4)
```
