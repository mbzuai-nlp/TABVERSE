```python
import json
from pathlib import Path

INPUT = "totto.json"
OUTPUT = "totto.cleaned.json"

def should_drop(block: dict, threshold: int = 4) -> bool:
    if not isinstance(block, dict):
        return False
    failures = sum(1 for v in block.values() if v == "CONNECTION_FAILED")
    return failures >= threshold

def clean_sample(sample: dict) -> dict:
    for k in ("html", "markdown", "latex"):
        if k in sample and should_drop(sample[k]):
            sample.pop(k)
    return sample

def clean_data(data):
    if isinstance(data, list):
        return [clean_sample(d) if isinstance(d, dict) else d for d in data]
    if isinstance(data, dict):
        return clean_sample(data)
    return data

# Load, clean, save
with open(INPUT, "r", encoding="utf-8") as f:
    data = json.load(f)

cleaned = clean_data(data)

with open(OUTPUT, "w", encoding="utf-8") as f:
    json.dump(cleaned, f, ensure_ascii=False, indent=4)

print(f"Saved cleaned dataset to {OUTPUT}")
```
