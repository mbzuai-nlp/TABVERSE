```python
import json
```

```python
output_path = "../data/3-suc/totto.json"
with open("../data/2-task/totto.json", "r") as f:
    data = json.load(f)
```

```python
len(data)
```

```python
# from data list only select those samples based on unique image_id, when duplicated image_id found, select the first one
unique_data = {}

for sample in data:
    image_id = sample["image_id"]
    if image_id not in unique_data:
        unique_data[image_id] = sample

len(unique_data)
```

```python
# change unique_data from dict to list
unique_data_list = list(unique_data.values())
```

```python
# remove query and label from each sample in unique_data_list
for sample in unique_data_list:
    if "query" in sample:
        del sample["query"]
    if "label" in sample:
        del sample["label"]
```

```python
unique_data_list[0]
```

```python
import random

for sample in unique_data_list:
    table = sample['table']
    header = table['header']
    rows = table['rows']
    n_rows = len(rows)
    n_cols = len(header)

    # Random selections
    rand_row_idx = random.randint(0, n_rows - 1)
    rand_col_idx = random.randint(0, n_cols - 1)
    cell_value = rows[rand_row_idx][rand_col_idx]

    row_retrieval_idx = random.randint(0, n_rows - 1)
    column_retrieval_idx = random.randint(0, n_cols - 1)

    suc = {
        "table_partition": f"{rows[0][0]}|{rows[-1][-1]}",
        "size_detection": f"{n_rows}|{n_cols}",
        "cell_value": cell_value,
        "cell_lookup": f"{rand_row_idx}|{rand_col_idx}",
        "reverse_lookup_indices": f"{rand_row_idx}|{rand_col_idx}",
        "reverse_lookup": cell_value,
        "column_idx": column_retrieval_idx,
        "column_retrieval": header[column_retrieval_idx],
        "row_idx": row_retrieval_idx,
        "row_retrieval": "|".join(rows[row_retrieval_idx]),
        "table_first_cell": rows[0][0],
        "table_last_cell": rows[-1][-1],
        "number_of_rows": n_rows,
        "number_of_columns": n_cols
    }

    sample['suc'] = suc
```

```python
unique_data_list[0]
```

```python
# reset id field to be sequential integers starting from 0
for idx, sample in enumerate(unique_data_list):
    sample['id'] = idx
```

```python
with open(output_path, 'w', encoding='utf-8') as file:
    json.dump(unique_data_list, file, ensure_ascii=False, indent=4)
```
